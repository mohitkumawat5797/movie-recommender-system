```python
def convert(obj):
    l = []
    for i in obj:
        l.append(i['name'])
    return l
```

```python
convert('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert(obj):
    l = []
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head(1)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head(1)
```

```python
movies['cast'][0]
```

```python
def convert3(obj):    
    l = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            l.append(i['name'])
            counter+=1
        else:
            break
    return l
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head(1)
```

```python
movies['crew'][0]
```

```python
def fetch_director(obj):
    l = []
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            l.append(i['name'])
            break
    return l
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies['overview'] = movies['overview'].fillna('')
movies['overview'] = movies['overview'].apply(lambda x: x.split())
#movies['overview'].apply(lambda x:x.split())
```

```python
movies.iloc[1]
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","")
                                  for i in x])
```

```python
movies['genres']
```

```python
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","")
                                  for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","")
                                  for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","")
                                  for i in x])
```

```python
movies.iloc[1]
```

```python
movies.head(1)
```

```python
movies['tags'] = movies['genres'] + movies['keywords'] + movies['overview'] + movies['cast'] + movies['crew']
```

```python
movies.head(1)
```

```python
new_df = movies[['id','title','tags']]
new_df.head()
```

```python
#tags ki list ko string me convert karna hai
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags'][2]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
new_df.head(1)
```

```python
#stemming karte hai
#Stemming is a text preprocessing technique in NLP where we reduce words to their root/base form.
#It cuts words down to a common base so similar words are treated the same.
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
ps.stem('adventurer')
```

```python
def stem(text):
    y =[]
    
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
    
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

```python
new_df['tags'][1]
```

```python
# vecorization karenge or stopwords ko sambhalna padega text me se 
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=2000,stop_words = 'english')
```

```python
cv.fit_transform(new_df['tags']).toarray().shape
```

```python
list(cv.get_feature_names_out())
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0][0]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse = True,key=lambda x:x[1])[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]]['title'])
```

```python
new_df.iloc[1214]['title']
```

```python
recommend('The Godfather')
```

```python
recommend('The Notebook')
```

```python
recommend('Batman Begins')
```
